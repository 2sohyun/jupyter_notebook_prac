## map, apply 함수

### 람다 함수를 사용하여 'Gender' 컬럼의 M은 Male로, F는 Female로 변경

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

s = df['Gender']

s.head()

# check = lambda x : x.replace('M','Male') if x=='M' else x.replace('F','Female')
# result = s.map(check)
result = s.map(lambda x: x.replace('M', 'Male') if x == 'M' else x.replace('F', 'Female'))
print(result)
```

### 딕셔너리를 사용하여 M은 Male로, F는 Female로 변경

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

s = df['Gender']

s.head()

# check = lambda x : x.replace('M','Male') if x=='M' else x.replace('F','Female')
# result = s.map(check)
result = s.map({'M':'Male', 'F':'Female'})
print(result)
```

### 새로운 계산법을 적용하여 Need_Diet 컬럼 내용을 변경
- Age가 30 이상이면 BMI에 1.1을 곱하고
- Gender가 M 이라면 다시 결과값에 1.1을 곱한다
- 최종 결과값이 20 이상이라면 Need_Diet를 True로 설정, 아니라면 False로 설정

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)
df['BMI'] = df['Weight']/df['Height']**2
print(df)

def myfunc(df):
    if df['Age'] >= 30 :
        df['BMI'] = df['BMI']*10000
    if df['Gender'] == 'M' :
        df['BMI'] = df['BMI']*10
    if df['BMI'] >= 20 :
        return True
    else:
        return False

df['Need_Diet'] = df.apply(myfunc, axis = 1)
print(df)
```

### 컬럼명을 모두 소문자로 바꾸기

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)
df['BMI'] = df['Weight']/df['Height']**2
print(df)

result = df.rename(columns = lambda x : x.lower())
print(result)
```

## Concat함수
axis=1 세로축 / axis=0 가로축 나누기

### DataFrame을 세로축으로 나누어 df1, df2에 할당하고, 다시 합하여 df3에 할당하기

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)
df['BMI'] = df['Weight']/df['Height']**2
print(df)

df1 = df.loc[:,:'is_student']
df2 = df.loc[:,'BMI':]

print(df1)
print(df2)

df3 = pd.concat([df1,df2], axis=1)

print(df3)
```

### DataFrame을 가로축으로 나누고, 다시 합하기

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)
df['BMI'] = df['Weight']/df['Height']**2
print(df)

df1 = df.loc[:2]
df2 = df.loc[3:]

print(df1)
print(df2)

df3 = pd.concat([df1,df2], axis=0)

print(df3)
```

### Index가 겹칠 경우2
ignore_index=True 옵션으로 인덱스가 겹치지 않도록 새로운 숫자 자동 생성됨

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)
df['BMI'] = df['Weight']/df['Height']**2
print(df)

df3 = pd.concat([df,df], axis=0, ignore_index=True)
print(df3)
```
