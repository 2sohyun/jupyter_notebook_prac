# 실습 데이터 베이스
## sqlite3
- 오픈소스 데이터베이스 SQLite의 파이썬 모듈
- lightweight disk-based database
- 독립적인 서버 프로세스를 가지고 있지 않다.(서버-클라이언트 모델x)
- 로컬에서만 사용

```python
import sqlite3

# 데이터베이스 생성(연결)
conn = sqlite3.connect('company.db')

try:
  # 커서(Cursor) 생성
  # 커서: 명령문(SQL) 실행, 결과의 현재 위치 표시
  cursor = conn.cursor()
    
  # 테이블 생성(CREATE), 같은 이름의 테이블이 이미 있는 경우 제거(DROP)
  cursor.execute("DROP TABLE IF EXISTS DEPARTMENT;")
  cursor.execute("CREATE TABLE DEPARTMENT(DID INT PRIMARY KEY, DNAME VARCHAR(32), FLOOR VARCHAR(32));")
    
  # 데이터 입력(INSERT)
  # sql 직접 입력
  cursor.execute("INSERT INTO DEPARTMENT VALUES(1, 'Finance', '10F');")
  cursor.execute("INSERT INTO DEPARTMENT VALUES(2, 'Sales', '9F');")
  
  # ? placeholder 사용
  cursor.execute("INSERT INTO DEPARTMENT VALUES(?, ?, ?);", (3, 'Research', '9F'))
  
  # name placeholder 사용
  cursor.execute("INSERT INTO DEPARTMENT VALUES(:f1, :f2, :f3)", {'f1':4, 'f2':'HR', 'f3':'10F'})

  # 트랜잭션 커밋 - 데이터베이스에 업데이트를 반영
  conn.commit()
  
  # 추가 데이터 입력
  cursor.execute("INSERT INTO DEPARTMENT VALUES(5, 'Marketing', '10F');")
  
  # 롤백- 최그 커밋 이후의 작업들을 취소
  conn.rollback()
  
  # 결과 확인
  # fetchall(): 쿼리의 결과를 모두 가져오기
  cursor.execute("SELECT * FROM DEPARTMENT")
  result = cursor.fetchall()
  for r in result:
    print(r)
  
except sqlite3.Error as e:
  # 데이터베이스 에러 처리
  print("sqlite error: ", e)

finally:
  # 데이터베이스 연결 해제
  conn.close()
```

## 쿼리 수행 메소드
- execute(sql, [파라미터들]) : 하나의 sql문 실행
- executemany(sql, seq_of_param) : 모든 매개변수 시퀀스나 매핑에 대하여 sql 실행
- executescript(sql_script) : 한 번에 여러 SQL문을 실행

```python
import sqlite3

# 데이터베이스 생성(연결)
conn = sqlite3.connect('company.db')

try:
  # 커서(Cursor) 생성
  cursor = conn.cursor()

  # 테이블 생성(CREATE), 같은 이름의 테이블이 이미 있는 경우 제거(DROP)
  cursor.execute("DROP TABLE IF EXISTS DEPARTMENT")
  cursor.execute("CREATE TABLE DEPARTMENT(DID INT, DNAME VARCHAR(32), FLOOR VARCHAR(32))")
    
  # 데이터 입력(INSERT)
  seq_of_param = [
      (1, 'Finance', '10F'),
      (2, 'Sales', '9F'),
      (3, 'Research', '9F'),
      (4, 'HR', '10F')
  ]
  sql = "INSERT INTO DEPARTMENT VALUES(?, ?, ?)"
  cursor.executemany(sql, seq_of_param)

  # 트랜잭션 커밋
  conn.commit()
  
  # 결과 확인
  result = cursor.execute("SELECT * FROM DEPARTMENT").fetchall()
  for r in result:
    print(r)
  
except sqlite3.Error as e:
  # 데이터베이스 에러 처리
  print("sqlite error: ", e)

finally:
  # 데이터베이스 연결 해제
  conn.close()
```

```python
import sqlite3

# 데이터베이스 생성(연결)
conn = sqlite3.connect('company.db')

try:
  # 커서(Cursor) 생성
  cursor = conn.cursor()
    
  # sql script 실행
  # 각 sql문은 ; 문자로 구분
  cursor.executescript("""
            DROP TABLE IF EXISTS DEPARTMENT;
            CREATE TABLE DEPARTMENT(DID INT, DNAME VARCHAR(32), FLOOR VARCHAR(32));
            
            INSERT INTO DEPARTMENT VALUES(1, 'Finance', '10F');
            INSERT INTO DEPARTMENT VALUES(2, 'Sales', '9F');
            INSERT INTO DEPARTMENT VALUES(3, 'Research', '9F');
            INSERT INTO DEPARTMENT VALUES(4, 'HR', '10F');
            """)

  # 트랜잭션 커밋
  conn.commit()
  
  # 결과 확인
  result = cursor.execute("SELECT * FROM DEPARTMENT").fetchall()
  for r in result:
    print(r)
    
except sqlite3.Error as e:
  # 데이터베이스 에러 처리
  print("sqlite error: ", e)

finally:
  # 데이터베이스 연결 해제
  conn.close()
```

## 결과 탐색 메소드
- fetchone() : 질의 결과 집합의 다음 행(row)을 하나씩 반환, 데이터가 더 없을 때 None을 반환
- fetchmany(size = cursor.arraysize) : size 개수 만큼의 결과 행(row)을 리스트로 반환, 종료시 None 반환
- fetchall() : 질의 결과에서 (남은) 모든 행(row)를 반환

```python
import sqlite3

conn = sqlite3.connect('company.db')

try: 
  cursor = conn.cursor() 
  cursor.execute("SELECT * FROM DEPARTMENT WHERE FLOOR = '9F'")

  r = cursor.fetchone()
  while r != None:
    print(r)
    r = cursor.fetchone()

except sqlite3.Error as e:
  print("sqlite error: ", e)
```

## 연습문제3) 연습문제 1과 같은 내용을 파이썬을 사용하여 수행
- INSERT문은 executemany() 함수 사용
- SELECT문 결과의 출력은 fetchone() 함수 사용

### 1. CREATE문을 사용하여 EMPLOYEE 테이블 생성

```python
import sqlite3

# 데이터베이스 연결
conn = sqlite3.connect('company.db')

try: 
  cursor = conn.cursor()

  # 테이블 생성(Create)
  cursor.execute("DROP TABLE IF EXISTS EMPLOYEE")
  cursor.execute("CREATE TABLE EMPLOYEE(EID INT, FNAME VARCHAR(32), LNAME VARCHAR(32), DID INT, GENDER CHAR, SALARY INT);")
    
  # 데이터 입력(Insert)
  values = [
      (1, 'Brad', 'Nash', 3, 'M', 30000),
      (2, 'Franklin', 'Smith', 2, 'M', 40000),
      (3, 'Nicole', 'Kim', 1, 'F', 25000),
      (4, 'Anna', 'Mills', 3, 'F', 43000),
      (5, 'Tim', 'Allen', 3, 'M', 38000),
      (6, 'Jennifer', 'Grey', 4, 'F', 35000),
      (7, 'Tony', 'Parker', 2, 'M', 25000),
      (8, 'James', 'Wong', 1, 'M', 45000)
  ]
  cursor.executemany("INSERT INTO EMPLOYEE VALUES(?, ?, ?, ?, ?, ?);", values)
    
  # 트랜잭션 커밋 - 데이터베이스에 업데이트를 반영
  conn.commit()

  # 결과 확인
  cursor.execute("SELECT FNAME, LNAME, SALARY FROM EMPLOYEE WHERE SALARY >= 30000;")
  r = cursor.fetchone()
  while r:
    print(r)
    r = cursor.fetchone()

except sqlite3.Error as e:
  print("sqlite error: ", e)
    
finally:
  conn.close()
```

### import Libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
from datetime import datetime
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

### import Data

```python
filename = 'data/building1retail.csv'
df = pd.read_csv(filename, index_col=[0], 
                 date_parser=lambda x: datetime.strptime(x, "%m/%d/%Y %H:%M"))
df.head()
```

### Fit Data to Linear Regression

```python
X = pd.DataFrame(df['OAT (F)'])
y = pd.DataFrame(df['Power (kW)'])

model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
    model.fit(X.iloc[train,:], y.iloc[train,:])
    score = model.score(X.iloc[test,:], y.iloc[test,:])
    scores.append(score)

print(scores)
```

### Add time of day

```python

X = pd.DataFrame(df['OAT (F)'])

var = 'tod'
X[var] = X.index.hour

# drop_first = True removes multi-collinearity
add_var = pd.get_dummies(X[var], prefix=var, drop_first=True)

# Add all the columns to the model data
X = X.join(add_var)

# Drop the original column that was expanded
X.drop(columns=[var], inplace=True)

X.head()
```

### Fit Data to Multiple Regression

```python
y = pd.DataFrame(df['Power (kW)'])

model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
    model.fit(X.iloc[train,:], y.iloc[train,:])
    scores.append(model.score(X.iloc[test,:], y.iloc[test,:]))

print(scores)
```
