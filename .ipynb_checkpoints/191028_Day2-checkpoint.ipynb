## 비가 오는 날, 안오는 날은 평균 관중수가 얼마나 차이날까?
- 시간 데이터 다루기(Datetime64)
- 두 DataFrame을 Join하기 (merge 함수)

### 고려사항
1. 야구경기는 보통 오후 5시 ~ 6시 사이에 시작된다 -> 15시~21시 사이의 강수량만 분석에 사용하자
2. 현재 날씨 데이터는 '시간별' 강수량을 나타낸다 -> groupby 함수를 사용하여, 해당 구장/날짜의 '총 강수량' 을 계산하자
3. Join key 선택 -> 하루에 한 구장에서는 한 경기만 열리므로, '구장' 과 '날짜' 를 Key로 하면, 각 경기를 구분할 수 있다.

### Step1) 시간 데이터 다루기 (예제 코드의 데이터 처리 과정)
1. df_weather에서 15시~21시 사이의 데이터만 추출
2. '구장','날짜'로 group을 만들어, 해당 구장/날짜의 총 강수량을 계산
3. '구장','날짜','강수량' 으로 이루어진 날씨 DataFrame을 새로 생성
4. '구장','날짜' 를 Join Key로 하여, df_baseball 과 Join

### Datetime 자료형의 property에 접근하기

```python
import datetime
# df_weather '일시' 컬럼의 1000번째 데이터를 datetime_value 변수에 할당

datetime_value = df_weather['일시'].loc[1000]
datetime_value
```

```python
import pandas as pd
import datetime

# datetime 자료형의 property에 접근 (단일 데이터)
# date()는 함수로 호출하는 것에 주의

print(" day: {}\n dayofyear: {}\n dayofweek: {}\n hour: {}\n date: {}\n".format(datetime_value.dt.day, datetime_value.dt.dayofyear, datetime_value.dt.dayofweek, datetime_value.dt.hour, datetime_value.dt.date))
```

```python
# Series단위로 datetime 자료형의 property에 접근
# Series단위로 datatime property에 접근하려면 series.dt.day 와 같이 접근
# date()는 함수로 호출하는 것에 주의

# 빈 데이터프레임 생성
df_result = pd.DataFrame()
# day, dayofyear, dayofweek, hour, date를 Series 단위로 생성하여, df_result에 컬럼 생성
df_result['일시'] = df_weather['일시']
df_result['day'] = df_weather['일시'].dt.day
df_result['dayofyear'] = df_weather['일시'].dt.dayofyear
df_result['dayofweek'] = df_weather['일시'].dt.dayofweek
df_result['hour'] = df_weather['일시'].dt.hour
df_result['date'] = df_weather['일시'].dt.date

# 결과 확인
df_result
```

### Step2) Group By
1. df_weather에서 15시~21시 사이의 데이터만 추출
2. '구장','날짜'로 group을 만들어, 해당 구장/날짜의 총 강수량을 계산
3. '구장','날짜','강수량' 으로 이루어진 날씨 DataFrame을 새로 생성
4. '구장','날짜' 를 Join Key로 하여, df_baseball 과 Join

```python
# 오후 15:00 ~ 20:59 까지의 데이터만 추출하여 df_weather_1521 변수에 할당

df_weather_1521 = df_weather[(df_weather['일시'].dt.hour>15) & (df_weather['일시'].dt.hour<20)]

df_weather_1521
```

```python
# Groupby를 이용하여 '구장','년월일 (date)' 별 총 강수량 구하기

# '년월일' 은 어떻게? -> df_weather_1521['일시'].dt.te
grouped = df_weather_1521.groupby([df_weather_1521['일시'].dt.date, df_weather_1521['구장']])                                                                                 
# 총 강수량 구하기
grouped_sum = grouped['강수량(mm)'].sum()

# 결과 확인하기
grouped_sum
```

### Step3) Join을 위한 구장/날짜별 강수량 DataFrame 생성
1. df_weather에서 15시~21시 사이의 데이터만 추출
2. '구장','날짜'로 group을 만들어, 해당 구장/날짜의 총 강수량을 계산
3. '구장','날짜','강수량' 으로 이루어진 날씨 DataFrame을 새로 생성
4. '구장','날짜' 를 Join Key로 하여, df_baseball 과 Join

```python
#grouped_sum의 Multi Index 제거
result = grouped_sum.reset_index()

df_rain_per_loc_day = result
df_rain_per_loc_day
```

```python
df_rain_per_loc_day = pd.DataFrame(df_rain_per_loc_day)
df_rain_per_loc_day.dtypes
```

### Join Key는 각 DataFrame의 컬럼명과, 각 컬럼의 자료형이 같아야 하므로
df_rain_per_loc_day '일시' 컬럼의 데이터 타입을 'datetime64'로 바꾸고
df_rain_per_loc_day '일시' 컬럼명을 '날짜' 로 바꾸자

```python
# result_rain의 '일시' 컬럼의 dtype을 'datetime64'로 바꾸기
df_rain_per_loc_day['일시'] = df_rain_per_loc_day['일시'].astype('datetime64')
df_rain_per_loc_day.dtypes
# result_rain의 '일시' 컬럼의 컬럼명을 '날짜'로 바꾸기
df_rain_per_loc_day = df_rain_per_loc_day.rename(columns = {"일시":"날짜"})
df_rain_per_loc_day
```

### Step4) Join 하고, 분석
1. df_weather에서 15시~21시 사이의 데이터만 추출
2. '구장','날짜'로 group을 만들어, 해당 구장/날짜의 총 강수량을 계산
3. '구장','날짜','강수량' 으로 이루어진 날씨 DataFrame을 새로 생성
4. '구장','날짜' 를 Join Key로 하여, df_baseball 과 df_rain_per_loc_day를 Join

### merge 함수
result = dataframe1.merge(dataframe2, on=[join keys], how='inner')

```python
# df_baseball 과 df_rain_per_loc_day 를 '구장'/'날짜' 를 Key로 inner join
result_merged = df_rain_per_loc_day.merge(df_baseball, on=['구장','날짜'], how='inner')

# 결과 확인
result_merged.head(1000)
```

### 분석예시 : 강수 여부를 기준으로 평균 관중수 비교

```python
# 강수여부를 기준으로 groupby 먼저 하기
result_merged['강수량(mm)'] = result_merged['강수량(mm)'].astype('float64')
result_merged['강수량(mm)'].dtypes
grouped = result_merged.groupby(['강수량(mm)'])
grouped
```

```python
# grouped를 이용하여 강수여부에 따른 평균 관중수 구하기(True : 비가 온 날, False : 비가 오지 않은 날)
grouped = result_merged.groupby(['강수량(mm)']).apply(lambda x: x['강수량(mm)']>0.0)
grouped
```
