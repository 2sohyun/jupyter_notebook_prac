## 데이터 선택 및 조작

### DataFrame의 size, shape, ndim

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

print(df.size)
#DataFrame 내의 총 데이터 개수 (row * column)
print(df.shape)
#row수, column수
print(df.ndim)
#데이터의 차원
```

### 'Name', 'Age', 'Gender' 컬럼 선택

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

result = df.loc[:,'Name':'Gender']
print(result)
result = df.loc[:,['Name','Age','Gender']]
print(result)
result = df.loc[0:4,['Name','Age','Gender']]
print(result)

#여러가지 방법 존재

result = df.loc[2]
print(result)
# 2번째 인덱스에 있는 값 리스트 출력
result = df.iloc[2]
print(result)
```

### 두 개의 Column 값을 사용하여, BMI 컬럼을 DataFrame에 추가
BMI = Weight / Height(미터) ^ 2

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2

print(df)
```

### BMI가 20이 넘는 사람은 True, 20 미만인 사람은 False를 값으로 갖는 'Need_Diet' 컬럼 추가


```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2

df['Need_Diet'] = df['BMI']>20

print(df)
```

## Column 삭제

### del을 이용하여 'Weight' Column 삭제

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2
df['Country'] = 'Korea'

del df['Weight']
print(df)
```

### pop을 이용하여 'Height' Column 삭제
s 변수에 지워진 'Height' Column이 할당된다.

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2
df['Country'] = 'Korea'

s = df.pop('Height')
# s 변수에 Height 컬럼이 할당되있는 거니까, df를 출력하면 Height를 뺀 값들이 출력
print(df)
```

### drop 함수를 이용하여 2~4번째 row 삭제 후 result 변수에 할당

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2
df['Country'] = 'Korea'

result = df.drop(range(1,4), axis=0)
print(result)
```

### drop 함수를 이용하여 BMI, Need_Diet Column 삭제 후 result 변수에 할당

```python
import pandas as pd

#DataFrame 생성
dict_list = {
    'Name' : ['Baek','Kim','Park','Han','Lee'],
    'Age' : [21,31,17,41,51],
    'Gender' : ['M','F','F','M','M'],
    'is_student' : [True, False, True, False, False]
}

df = pd.DataFrame(dict_list)
print(df)

# 1차원 데이터를 이용하여 Weight 컬럼을 DataFrame에 추가
df['Weight'] = [50,55,60,65,70]
#df['Weight'] = (50,55,60,65,70)
#df['Weight'] = pd.Series([50,55,60,65,70])
'''
df['Weight'] = {
    0: 50,
    1: 55,
    2: 60,
    3: 65,
    4: 70
}
'''
df['Height'] = (155,180,193,153,166)

df['BMI'] = df['Weight']/df['Height']**2
df['Country'] = 'Korea'

result = df.drop(["BMI","Height"], axis=1)
print(result)
```
