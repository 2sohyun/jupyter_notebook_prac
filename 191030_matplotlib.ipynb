### [개념10] 가로 축, 세로 축 이동

- ax.spines['top'].set_color('none')

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(X), np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, C, c="blue", linewidth=4, linestyle="--")
ax.plot(X, S, c="red", linewidth=1, linestyle="-")
ax.set_xlim(X.min() * 1.5, X.max() * 1.5)
ax.set_ylim(C.min() * 1.5, C.max() * 1.5)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, +1])
ax.set_xticklabels([r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$+\frac{\pi}{2}$', r'$+\pi$'])
ax.set_yticklabels(["-1", "0", "+1"])

ax.spines['right'].set_color('y')
ax.spines['top'].set_color('g')
# ax.xaxis.set_ticks_position('top')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('right')
ax.spines['left'].set_position(('data',0))

#기본적으로 xaxis는 bottom, yaxis는 left으로 설정되어있다.

plt.show()
```

### [개념11] 범례 표시하기
- ax.legend()를 이용해 그래프의 범례를 표시할 수 있다.
- 단, 그 전에 ax.plot(), ax.scatter()에 label 파라미터를 설정하는 것이 좋다.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(X), np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, C, c="blue", linewidth=4, linestyle="--", label="cosine")
ax.plot(X, S, c="red", linewidth=1, linestyle="-", label="sine")
# label 파라미터 설정

ax.set_xlim(X.min() * 1.5, X.max() * 1.5)
ax.set_ylim(C.min() * 1.5, C.max() * 1.5)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, +1])
ax.set_xticklabels([r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$+\frac{\pi}{2}$', r'$+\pi$'])
ax.set_yticklabels(["-1", "0", "+1"])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

ax.legend(loc='upper left')
# 그래프의 범례 표시

plt.show()
```

### [개념13] 특징점 강조
- ax.scatter()와 ax.annotate()를 이용해 그래프의 특정점을 강조

```python
# x=2π/3 인 점을 강조

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(X), np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, C, c="blue", linewidth=4, linestyle="--", label="cosine")
ax.plot(X, S, c="red", linewidth=1, linestyle="-", label="sine")
ax.set_xlim(X.min() * 1.5, X.max() * 1.5)
ax.set_ylim(C.min() * 1.5, C.max() * 1.5)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, +1])
ax.set_xticklabels([r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$+\frac{\pi}{2}$', r'$+\pi$'])
ax.set_yticklabels(["-1", "0", "+1"])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

ax.legend(loc='upper left')
ax.grid(True)

t = 2 * np.pi / 3 # 3분에 2파이
ax.plot([t, t], [0, np.cos(t)], c='blue', linewidth=1, linestyle=":")
ax.scatter(t, np.cos(t), 50, c='green')
#ax.annotate(r'$cos(\frac{2\pi}{3} )=-\frac{1} {2} $', xy=(t, np.cos(t)))
ax.annotate("comment1", xy=(t, np.cos(t)))
ax.annotate(r'$sin(\frac{2\pi}{3} )=-\frac{1} {2} $', xy=(t, np.sin(t)))
ax.plot([t, t],[0, np.sin(t)], c='red', linewidth=1, linestyle=":")
ax.scatter(t, np.sin(t), 50, c='green')

plt.show()
```
