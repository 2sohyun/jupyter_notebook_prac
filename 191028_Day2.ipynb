## 비가 오는 날, 안오는 날은 평균 관중수가 얼마나 차이날까?
- 시간 데이터 다루기(Datetime64)
- 두 DataFrame을 Join하기 (merge 함수)

### 고려사항
1. 야구경기는 보통 오후 5시 ~ 6시 사이에 시작된다 -> 15시~21시 사이의 강수량만 분석에 사용하자
2. 현재 날씨 데이터는 '시간별' 강수량을 나타낸다 -> groupby 함수를 사용하여, 해당 구장/날짜의 '총 강수량' 을 계산하자
3. Join key 선택 -> 하루에 한 구장에서는 한 경기만 열리므로, '구장' 과 '날짜' 를 Key로 하면, 각 경기를 구분할 수 있다.

### 시간 데이터 다루기 (예제 코드의 데이터 처리 과정)
1. df_weather에서 15시~21시 사이의 데이터만 추출
2. '구장','날짜'로 group을 만들어, 해당 구장/날짜의 총 강수량을 계산
3. '구장','날짜','강수량' 으로 이루어진 날씨 DataFrame을 새로 생성
4. '구장','날짜' 를 Join Key로 하여, df_baseball 과 Join
