## Enumerate함수

```python
colors = ['yellow', 'orange', 'red', 'plum', 'pink', 'green', 'blue', 'black']
for i in range(0, len(colors)):
    print(colors[i]) 
#인덱스를 이용하는 접근방식
for color in colors:
    print(color)
#인덱스를 이용하지 않는 접근방식

for i, color in enumerate(colors):
    print("{}:{}".format(i,color))
#enumerate(LIST)함수 : 열거하다 인덱스:리스트로 출력

names = ['철수', '영희', '영수']
for i, name in enumerate(names):
    print('{}번: {}'.format(i+1,name))
#for문처럼 반복되는 구간에서 객체가 현재 어느 위치에 있는 지 알려주는 인덱스 값이 필요할 때 유용한 함수
#i+1로 설정함으로써 기본 인덱스 0부터가 아닌 1부터 출력
```

## 실습1] LIST 문제 - enumerate(LIST) 함수 활용
문자열을 담고 있는 리스트 colors에서 "p"로 시작하는 원소들의 index를 출력하라.

```python
colors = ["red", "blue", "green", "black", "white"]
colors.append("pink")
colors.extend(["yellow","plum"])

for i, color in enumerate(colors):
    if color.startswith("p"):
        print("index: {} & color: {}".format(i,color))
    
```

## 실습2] LIST 문제 
색깔에 대한 선호도를 조사하여 colors와 counts의 두 리스트에 동일한 순서로 저장하였다. 두 리스트를 선언하고, 색깔별로 좋아하는 사람의 숫자를 출력하라.

```python
colors = ["yellow", "white", "red", "plum", "pink", "blue", "black"]
counts = [54, 31, 100, 87, 32, 98, 5]

for i in range(len(colors)):
    print("{} 색깔을 좋아하는 사람의 수는 {}명이다.".format(colors[i],counts[i]))
```

## 실습3] LIST 문제 - zip() 함수 사용
색깔에 대한 선호도를 조사하여 colors와 counts의 두 리스트에 동일한 순서로 저장하였다. 두 리스트를 선언하고, 색깔별로 좋아하는 사람의 숫자를 출력하라.

```python
colors = ["yellow", "white", "red", "plum", "pink", "blue", "black"]
counts = [54, 31, 100, 87, 32, 98, 5]

for color,count in zip(colors,counts):
    print("{} 색깔을 좋아하는 사람의 수는 {}명이다.".format(color,count))
#zip(): 동일한 갯수의 요소 값을 갖는 시퀀스 자료형을 묶어주는 역할(2개 이상의 순회형 자료를 받아 순서대로 묶어주기)
```

## 실습5] map(function, iterable) 사용할 경우
숫자(1,3)와 문자열로 작성된("2")가 공존하는 리스트를 받아 리스트 안의 모든 자료를 숫자로 변환하는 프로그램

## 실습9] filter 활용
0 이상 200 이하의 숫자 중 임의로 20개를 뽑아 리스트에 저장하고 이 리스트 중에서 3의 배수만 뽑아내는 코드를 filter()를 이용해 작성

```python
import random

rList = random.sample(range(0,201),20)
print(rList)
a = list(filter(lambda a: a%3 == 0, rList))
print(a)
```

## 실습10] 성능에 대한 고민
각각 5만개의 숫자를 포함하는 두 리스트에 공통으로 존재하는 숫자의 개수를 세어보기

```python

%%time

import random

pool = range(0, 10_000_000)

sample1 = random.sample(pool, 50_000)
sample2 = random.sample(pool, 50_000)

sample_1L = list(sample1)
sample_2L = list(sample2)

count = 0
for e1 in sample_1L:
    if e1 in sample_2L:
        count += 1
    else:
        print ("{:,}개의 공통 원소가 있습니다.".format(count))
```

## numpy로 보다 편리한 array 연산 사용
